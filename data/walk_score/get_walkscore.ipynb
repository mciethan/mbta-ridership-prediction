```python
import pandas as pd
import requests

rapid_trans_facilities = pd.read_csv('../stop_geometry/rapid_transit_facilities.csv')

# Replace 'YOUR_API_KEY' with your actual Walk Score API key
API_KEY = 'b12d2c039749ebd2e8f67ffe0d9f2b83'

# Function to get Walk Score for a given latitude and longitude
def get_walk_score(lat, lon):
    url = 'https://api.walkscore.com/score'
    params = {
        'format': 'json',
        'lat': lat,
        'lon': lon,
        'transit': 1,
        'bike': 1,
        'wsapikey': API_KEY
    }
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        data = response.json()
        if data['status'] == 1:  # Successful response
            return data['walkscore']
        else:
            return None  # If no score is available or other statuses
    except Exception as e:
        print(f"Error retrieving Walk Score for ({lat}, {lon}): {e}")
        return None

# Apply the function to each row and store results in a new 'walk_score' column
rapid_trans_facilities['walk_score'] = rapid_trans_facilities.apply(lambda row: get_walk_score(row['stop_lat'], row['stop_lon']), axis=1)

rapid_trans_facilities
```
