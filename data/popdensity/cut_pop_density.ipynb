```python
import pandas as pd
import geopandas as gpd

dtype_1 = {"w_geocode":str}

dtype_2 = {"GEOID20":str}

total_pop = gpd.read_file("acs2022_5yr_B01003_15000US250235401013/acs2022_5yr_B01003_15000US250235401013.shp",dtype=dtype_1).to_crs("EPSG:26986")
```

```python
total_pop
```

```python
total_pop["area"] = total_pop['geometry'].area
total_pop.rename(columns={"B01003001":"tot_pop"},inplace=True)
```

```python
total_pop["ppsqft"] = total_pop["tot_pop"] / total_pop["area"]
total_pop.to_crs("EPSG:4326",inplace=True)
```

```python
station_areas = pd.read_csv("../stations_routes_f23_geoms_nonoverlapping.csv")

station_areas= gpd.GeoDataFrame(station_areas,geometry=gpd.GeoSeries.from_wkt(station_areas["walk_poly_wkt"]))
station_areas = station_areas.set_crs("EPSG:4326")
```

```python
total_pop['centroid'] = total_pop.centroid

# Step 2: Create a new GeoDataFrame for centroids with the same CRS
centroids_gdf = gpd.GeoDataFrame(total_pop, geometry='centroid', crs=total_pop.crs)
```

```python
centroids_gdf
```

```python
stations_w_pop = gpd.sjoin(station_areas,centroids_gdf, predicate='contains')
```

```python
stations_w_pop
```

```python
stations_pop = stations_w_pop.groupby(["route_id","parent_station"]).mean()[["ppsqft"]].reset_index()

stations_pop["tot_pop"] = stations_w_pop.groupby(["route_id","parent_station"]).sum()[["tot_pop"]].reset_index()["tot_pop"]
```

```python
stations_pop
```

```python
stations_pop.to_csv("stations_routes_acs2022_5yr_pop.csv")
```
